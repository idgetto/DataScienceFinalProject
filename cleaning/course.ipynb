```python
def sortdata(columnx, columny, listx = []):
    """
    if columnx , return all cases in columny as a dictionary
    columnx: name of the column we want to use to get the key
    columny: name of the column we want to use to get the values
    listx: if listx is not given, all unique values in columnx are used as keys
           if listx is given, only the values in listx are used as keys
    """
    if len(listx) == 0:
        listx = data[columnx].unique()
        listx.sort()
    dicty = {}
    for x in listx:
        dicty[x] = []
    for i in data.index:
        x = data[columnx][i]
        y = data[columny][i]
        if y in dicty[x]:
            continue
        else:
            dicty[x].append(y)
    return dicty

coursecoded = sortdata('course-sectiontitle','code')
codecoursed = sortdata('code','course-sectiontitle')
```

```python
for key in coursecoded.keys():
    if len(coursecoded[key]) > 1:
        print key + ': ' + str(coursecoded[key])
```
