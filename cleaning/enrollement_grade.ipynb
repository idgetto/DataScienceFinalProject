```python
import pandas as pd

data = pd.read_csv('../data/registrations.csv')
data.head()
```

```python
data.grade.unique()
```

```python
gds = data.groupby('grade')
for gd in gds:
    print gd[0]
    print len(gd[1])
```

```python
import string

course_names = data['code'].unique()
course_names = [string.strip(course_name) for course_name in course_names]
df = pd.DataFrame(columns=course_names)
df.head()
```

```python
for gd in gds:
    if gd[0] == 'FR':
        fr = gd[1]
    elif gd[0] == 'SO':
        so = gd[1]
    elif gd[0] == 'JR':
        jr = gd[1]
    elif gd[0] == 'SR':
        sr = gd[1]
    so = so.append(fr)
    jr = jr.append(so)
    sr = sr.append(jr)
```

```python
grouped = data.groupby('studentid')
for group in grouped:
    student_id = group[0]
    registrations = group[1]
    names = registrations['code'].unique()
    names = [string.strip(name) for name in names]
    df.loc[student_id] = False
    df.loc[student_id, names] = True
    
df['studentid'] = df.index
```

```python
df.to_csv('../data/all_enrollment.csv', index=False)
```
