```python
import pandas as pd
data = pd.read_csv('../data/olin0216.csv')

data.head()
```

```python
string_cols = list(data.columns.values)
string_cols.remove('studentid')
print string_cols
```

```python
import string

def strip_data(series):
    res = []
    for data in series:
        if type(data) != float:
            res.append(string.strip(data))
        else:
            res.append('')
    return res

data[string_cols] = data[string_cols].apply(strip_data)
data.head()
```

```python
data.to_csv('../data/registrations.csv', index=False)
```
