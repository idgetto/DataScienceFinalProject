# Course_type

The course_type csv will add to the original dataset what general area the course is in. Each course will be assigned to categories such as 'design', 'electrical', 'mechanical', 'math' and such. This process was done manually, as there are no discrete classification on courses, and we wanted the students to get a feel for how the students classify courses at Olin.

```python
# Import pandas and load original dataset
import pandas as pd
data = pd.read_csv('../data/registrations.csv')
```

```python
data.head()
```

```python
# we generate a function to sort the csv into convenient portions to assign categories
def sortdata(columnx, columny, listx = []):
    """
    if columnx , return all cases in columny as a dictionary
    columnx: name of the column we want to use to get the key
    columny: name of the column we want to use to get the values
    listx: if listx is not given, all unique values in columnx are used as keys
           if listx is given, only the values in listx are used as keys
    """
    if len(listx) == 0:
        listx = data[columnx].unique()
        listx.sort()
    dicty = {}
    for x in listx:
        dicty[x] = []
    for i in data.index:
        x = data[columnx][i]
        y = data[columny][i]
        if y in dicty[x]:
            continue
        else:
            dicty[x].append(y)
    return dicty

coursecoded = sortdata('course-sectiontitle','code')
codecoursed = sortdata('code','course-sectiontitle')
```

```python
# observing the output from the function above
codecoursed
```

```python
# merging together the course code and the course name
for key in codecoursed.keys():
    if len(codecoursed[key]) > 1:
        print key + ': ' + str(codecoursed[key])
```

```python
import collections
d = collections.OrderedDict()

for key in codecoursed.keys():
    d[key] = str(codecoursed[key])
```

```python
# Saving a new csv with the code list, with each course type yet to be written in manually
course = pd.DataFrame({'code': d.keys(), 'coursetype': d.values()})
course.head
course.to_csv('../data/course_raw.csv', index=range(len(d)))
```
