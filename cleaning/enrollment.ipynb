```python
import pandas as pd

data = pd.read_csv('../data/registrations.csv')
data.head()
```

```python
import string

course_names = data['course-sectiontitle'].unique()
course_names = [string.strip(course_name) for course_name in course_names]
df = pd.DataFrame(columns=course_names)
df.head()
```

```python
grouped = data.groupby('studentid')
for group in grouped:
    student_id = group[0]
    registrations = group[1]
    names = registrations['course-sectiontitle'].unique()
    names = [string.strip(name) for name in names]
    df.loc[student_id] = False
    df.loc[student_id, names] = True
    
df['studentid'] = df.index
```

```python
df.head()
```

```python
df.to_csv('../data/enrollment.csv', index=False)
```
