Generating spider web plots for all courses for putting them up on the website

```python
import spider_web as sw

import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('../data/course-semesters.csv')
df.head()
```

```python
import re
import spider_web as sw

i = 0

grouped = df.groupby('name')
for (name, data) in grouped:
    all_ids = []
    id_lists = data.student_ids.values
    for id_list in id_lists:
        converted = re.split('[^\d]+', id_list)
        all_ids += converted[1:-1]
    
    safe_name = re.sub('[ \/]+', '_', name)
    print safe_name
    all_ids = [int(sid) for sid in all_ids]
    
    plt.title(name, y=1.2)
    plt.subplots_adjust(top=0.8)
    sw.students_spider_plot(all_ids)
    
    base_name = '../../OlinDataScienceCourseFinder.github.io/images/courses/spider/'
    file_name = base_name + safe_name
    plt.savefig(file_name)
    plt.clf()
```
