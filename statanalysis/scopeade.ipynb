```python
ade = ['ENGR4199', 'ENGR4290']
scope = ['ENGR4190']
adescoped = {'semester':[],
             'ADE':[],
             'SCOPE':[]}

scopesemesterlist = []
adesemesterlist = []

for i in cs.index:
    if cs.code[i] in scope:
        if cs.semester[i] not in scopesemesterlist:
            scopesemesterlist.append(cs.semester[i])

for i in cs.index:
    if cs.code[i] in ade:
        if cs.semester[i] not in adesemesterlist:
            adesemesterlist.append(cs.semester[i])

adescoped['semester'] = scopesemesterlist

for sem in range(len(scopesemesterlist)-len(adesemesterlist)):
    adescoped['ADE'].append(0)

for i in cs.index:
    if cs.code[i] in ade:
        for j in range(len(scopesemesterlist)):
            if cs.semester[i] == scopesemesterlist[j]:
                adescoped['ADE'].append(len(cs.student_ids[i][1:-1].split(',')))
                
for i in cs.index:
    if cs.code[i] in scope:
        for j in range(len(scopesemesterlist)):
            if cs.semester[i] == scopesemesterlist[j]:
                adescoped['SCOPE'].append(len(cs.student_ids[i][1:-1].split(',')))   
                
print adescoped
adescope = pd.DataFrame(adescoped)
```

```python
olin = pd.read_csv('../data/olin0216.csv')
olin.head()
```

```python
scope = ['ENGR4190']
ade = ['ENGR4199', 'ENGR4290']

def ignore_nan(inputs):
    """returns empty string if the value is NaN"""
    if type(inputs) == float:
        return ''
    else:
        return inputs
    
olin['majorconc'] = olin['major'] + ' ' + olin['concentration'].apply(ignore_nan)

ademajor = []

codes = olin.groupby('code')
for code in codes:
    if code[0] in ade:
        ademajor += code[1].majorconc.tolist()
```

```python
ades = Series(ademajor)
aded = ades.value_counts()
print aded
```

```python
ades.count()
```

```python
plt.pie(ades.value_counts(), autopct='%1.1f%%')
```

```python
scopemajor = []

codes = olin.groupby('code')
for code in codes:
    if code[0] in scope:
        scopemajor += code[1].majorconc.tolist()
        
scopes = Series(scopemajor)
scoped = scopes.value_counts()
print scoped
```

```python
scopes.count()
```

```python
plt.pie(scopes.value_counts(), autopct='%1.1f%%')
```

```python
allmajor = []
final = ade + scope

codes = olin.groupby('code')
for code in codes:
    if code[0] in final:
        allmajor += code[1].majorconc.tolist()
        
allm = Series(allmajor)
alld = allm.value_counts()
print alld
```

```python
allm.count()
```

```python
import collections

plt.figure(figsize=(10,10))
plt.barh(range(len(scoped)), scoped, color = 'blue', align='center', alpha = 0.5)


scopeda = collections.OrderedDict()
for key in scoped.keys():
    scopeda[key] = int(float(alld[key])/allm.count()*scopes.count())
for key in alld.keys():
    if key not in scopeda.keys():
        scopeda[key] = 0

plt.barh(range(len(scopeda)), scopeda.values(), color = 'red', align='center', alpha = 0.3)
plt.yticks(range(len(scopeda)), scopeda.keys())
```

Blue shows the actual number of students taking scope

Red shows the students that should be taking scope given the total number of students taking SCOPE and the distribution of students across major

```python
plt.figure(figsize=(10,10))
plt.barh(range(len(aded)), aded, color = 'blue', align='center', alpha = 0.5)


adeda = collections.OrderedDict()
for key in aded.keys():
    adeda[key] = int(float(alld[key])/allm.count()*ades.count())
for key in alld.keys():
    if key not in adeda.keys():
        adeda[key] = 0

plt.barh(range(len(adeda)), adeda.values(), color = 'red', align='center', alpha = 0.3)
plt.yticks(range(len(adeda)), adeda.keys())
```

blue shows the number of students taken ADE while red shows the number of student who should have taken the course based on total number of students taking ADE and the distribution of student taking each major

A lot of ECE does not take ADE while a lot of ME, Design E and Self designed E takes ADE. This might be due to the nature of SCOPE - SCOPE is more relevant for ECE than ADE
