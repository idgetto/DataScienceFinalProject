```python
def extract_semester(semester):
    """
    semester: 0304FA (unicode)
    year: 03 (unicode)
    fs: FA or SP (unicode)
    """
    fs = semester[-2:] #fall/spring
    if fs == 'FA':
        year = semester[0:2]
    elif fs == 'SP':
        year = semester[2:4]
    else:
        print 'unknown semester'
    return year, fs
```

### Creating new columns
year, term, course_fullname, major_fullname, code-semester

```python
"""
Creating new columns 
year: the year the course was taken
term: the term the semester was taken
"""

def semester_year(semester):
    fs = semester[-2:]
    if fs == 'FA':
        year = semester[0:2]
    elif fs == 'SP':
        year = semester[2:4]
    return int(year)

data['term'] = data['semester'].str[-2:]
data['year'] = data['semester'].apply(semester_year)
data.head()
```

```python
"""
Creating new column to join the name course_main and course_sub into course_fullname
"""
import numpy as np

def ignore_nan(course_sub):
    """returns empty string if the value is NaN"""
    if type(course_sub) == float:
        return ''
    else:
        return course_sub

data['course_fullname'] = data['course_main'] + ' ' + data['course_sub'].apply(ignore_nan)
data['major_fullname'] = data['major_main'] + ' ' + data['major_sub'].apply(ignore_nan)
data['code-semester'] = data['code'] + ' ' + data['semester']
data.head()
```

```python
def sortdata(columnx, columny, listx = []):
    """
    if columnx , return all cases in columny as a dictionary
    columnx: name of the column we want to use to get the key
    columny: name of the column we want to use to get the values
    listx: if listx is not given, all unique values in columnx are used as keys
           if listx is given, only the values in listx are used as keys
    """
    if len(listx) == 0:
        listx = data[columnx].unique()
        listx.sort()
    dicty = {}
    for x in listx:
        dicty[x] = []
    for i in data.index:
        x = data[columnx][i]
        y = data[columny][i]
        if y in dicty[x]:
            continue
        else:
            dicty[x].append(y)
    return dicty

# coursecodedict = sortdata('code', 'course_main', coursecodelist)
# coursedict = sortdata('course_main', 'code', courselist)
# testdict = sortdata('code', 'course_main')
```
